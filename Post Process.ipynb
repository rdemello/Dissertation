```python
thisPath = "Projections/Grouped"

test = ["United Kingdom"]
top30Countries = ["China", "United States","India", "Russian Federation", "Japan", 
                  "Canada", "Germany", "Korea, Rep.", "Brazil", "France", "Saudi Arabia", 
                  "United Kingdom", "Pakistan", "Mexico", "Iran, Islamic Rep.", "Turkey", "Italy", "Spain", 
                  "Indonesia", "Australia", "South Africa", "Vietnam", "Egypt, Arab Rep.", 
                  "Thailand", "Argentina","Nigeria","Poland", "Malaysia", "Venezuela, RB", "Congo, Dem. Rep."]

def getCountry(thisPath, country):
    thisCountry = pd.read_csv('{}/{}/Summary.csv'.format(thisPath, country))
    percent_change = ((thisCountry["2050"] - thisCountry["2020"])/thisCountry["2020"])*100
    absolute_change = thisCountry["2050"] - thisCountry["2020"]
    endResult = thisCountry["2050"]
    return endResult

def summaryTable(thisPath):
    newNP = np.zeros((1,1))
    mainResults = pd.DataFrame(data=newNP[1:,1:])
        
    for x, country in enumerate(top30Countries):
    
        thisCountryResults = getCountry(thisPath, country)
        mainResults = pd.concat([mainResults, thisCountryResults], axis=1)
    
    mainResults.columns= top30Countries
    mainResults["new"] = ["SSP1 Lower", "SSP1 Mean","SSP1 Upper","SSP2 Lower", "SSP2 Mean","SSP2 Upper",
                              "SSP3 Lower", "SSP3 Mean","SSP3 Upper","SSP4 Lower", "SSP4 Mean","SSP4 Upper",
                              "SSP5 Lower", "SSP5 Mean","SSP5 Upper",]
    mainResults = mainResults.set_index("new")
    
    mainResults.to_csv('{}/end_summary.csv'.format(thisPath))
    
    return mainResults
    
summaryTable(thisPath)
```

```python
def getAllYears(thisPath, country):
    thisCountry = pd.read_csv('{}/{}/Summary.csv'.format(thisPath, country))
    means = thisCountry.drop([0,2,3,5,6,8,9,11,12,14])
    return means
    
def summaryEndPoints(thisPath):
    
    for x in [0,1,2,3,4]:
        
        newNP = np.zeros((1,1))
        mainResults = pd.DataFrame(data=newNP[1:,1:])

        for country in top30Countries:
        
            allMeans = getAllYears(thisPath, country)
            thisMeans = allMeans.iloc[[x]]
            mainResults = pd.concat([mainResults, thisMeans])
            
            del allMeans, thisMeans
            
        mainResults["Unnamed: 0"]=top30Countries
        mainResults.to_csv('{}/SSP{}_Spread.csv'.format(thisPath, x+1))
        
summaryEndPoints(thisPath)
    
```
